### nullをそのままにして相関係数を求める

```python
# Nanを再び置き換える

# all_df['OCCUPATION_TYPE'] = all_df['OCCUPATION_TYPE'].replace(all_df['OCCUPATION_TYPE'].max(), np.nan)
# all_df[cate_c].head()
print('変換前のnull\n', all_df[cate_c].isnull().sum())


def label_encoding_ex_null(df_include_null):
    le = LabelEncoder()
    le_count = 0
    # Iterate through the columns
    for col in df_include_null:
        if df_include_null[col].dtype == 'object':
            # If 2 or fewer unique categories
            # if len(list(df_include_null[col].unique())) <= 2:
            # Train on the training data
            le.fit(df_include_null[col])
            # Transform both training and testing data
            df_include_null[col] = le.transform(df_include_null[col])
            if col in in_null_c:
                all_df[col] = all_df[col].replace(all_df[col].max(), np.nan)

            # Keep track of how many columns were label encoded
            le_count += 1
    
    print('%d columns were label encoded.' % le_count)

label_encoding_ex_null(all_df)
print('変換後のnull\n', all_df[cate_c].isnull().sum())
```

```python
# 欠損値を他の変数から予想できないか考える
# Find correlations with the target and sort
correlations = all_df.corr()
for col in in_null_c:
    # correlations = all_df.corr()

    # Display correlations
    print('Most Positive Correlations:\n', correlations[col].sort_values().tail(15))
    print('------', col, '----------')
    print('Most Negative Correlations:\n', correlations[col].sort_values().head(15))
    print("")
```
