```python
# ライブラリの読み込み
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math

import warnings
warnings.filterwarnings('ignore')

from sklearn.preprocessing import LabelEncoder
```

```python
# データの読み込み
# INPUT_DIRにtrain.csvなどのデータを置いているディレクトリを指定してください。
INPUT_DIR = "/content/drive/MyDrive/Colab Notebooks/02.（公開）コンペ2/input/"

train_df = pd.read_csv(INPUT_DIR + "train.csv")
test_df = pd.read_csv(INPUT_DIR + "test.csv")
sample_sub = pd.read_csv(INPUT_DIR + "sample_submission.csv")
```

```python
all_df = pd.concat([train_df, test_df], axis=0) # 訓練データとテストデータ結合

int_c = set(all_df.select_dtypes('int64').columns)
float_c = set(all_df.select_dtypes('float64').columns)
cate_c = set(all_df.select_dtypes('object').columns)

num_c = int_c | float_c
print(len(all_df.columns) == len(num_c | cate_c)) # すべてのカラムが選択できているか確認

num_c = list(num_c)
cate_c = list(cate_c)
print(len(num_c), len(cate_c))
```

#### 数値変数

```python
def show_num_x(column_list, fig_col = 4):
    df_col = 0
    fig_row = math.ceil(len(column_list) / fig_col)

    fig, axes = plt.subplots(nrows=fig_row, ncols=fig_col, figsize=(20, fig_row*4))
    fig.subplots_adjust(hspace=0.2)


    for i in range(fig_row):
        for j in range(fig_col):
            try:
                sns.distplot(x = all_df[column_list[df_col]], ax=axes[i, j])
                n_null = all_df[column_list[df_col]].isnull().sum()
                n_uniqe = all_df[column_list[df_col]].nunique()
                axes[i, j].set_title("%s (null:%d, unique:%d)"%(column_list[df_col], n_null, n_uniqe))
                df_col += 1
            except IndexError:
                print('IndexError')
```

```python
show_num_x(num_c)
```
